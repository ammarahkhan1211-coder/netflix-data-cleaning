```python
file_path = r"C:\Users\ammar\OneDrive\Desktop\netflix_titles.csv"
df = pd.read_csv(file_path)
df.head()
```

```python
print(df.isnull().sum())
```

```python
# List of columns to fill missing values
missing_text_columns = ['director', 'cast', 'country', 'rating', 'date_added', 'duration']

# Fill missing values with "unknown"
for col in missing_text_columns:
    if col in df.columns:  # make sure the column exists
        df[col] = df[col].fillna('unknown')

# Verify that missing values are handled
print("Missing values after filling:")
print(df.isnull().sum())
```

```python
# Clean column names: lowercase and replace spaces with underscores
df.columns = df.columns.str.lower().str.replace(' ', '_')

# Check the cleaned column names
print("Cleaned column names:")
print(df.columns)
```

```python
if 'date_added' in df.columns:
    df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')
```

```python
# Split genres into a list
df['genres_list'] = df['listed_in'].str.split(', ')
# Preview the first 5 rows
df[['title', 'listed_in', 'genres_list']].head()
```

```python
# First, get all unique genres
all_genres = set()  # empty set to avoid duplicates

# Go through each row and add genres to the set
for genres in df['genres_list']:
    all_genres.update(genres)  # add all genres from this row

# Create a new column for each genre
for genre in all_genres:
    df[genre] = df['genres_list'].apply(lambda x: 1 if genre in x else 0)

# Preview the first 5 rows
df.head(8)[['title'] + list(all_genres)[:10]]  # show title + first 5 genre columns
```

```python
# Define a function to categorize ratings
def categorize_rating(rating):
    if rating in ['g', 'pg', 'pg-13']:
        return 'family'
    elif rating in ['r', 'nc-17', 'tv-ma']:
        return 'adult'
    else:
        return 'unknown'

# Apply the function to create a new column
df['rating_category'] = df['rating'].apply(categorize_rating)

# Preview the first 10 rows
df[['title', 'rating', 'rating_category']].head(10)
```
